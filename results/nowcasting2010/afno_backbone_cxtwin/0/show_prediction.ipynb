```python
import h5py 
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
# load the data
f = h5py.File('prediction.h5', 'r')
print(f.keys())
input = f['input']
prediction = f['prediction']
```

```python
plt.figure(figsize=(12, 6), dpi=300)
vmax = np.max(input[0, -1])
vmin = np.min(input[0, -1])
plt.subplot(1, 2, 1)
plt.pcolor(input[1, -1], vmax=vmax, vmin=vmin)
plt.colorbar()

plt.subplot(1, 2, 2)
plt.pcolor(prediction[0, -1], vmax=vmax, vmin=vmin)
plt.colorbar()
```

```python
plt.figure(figsize=(12, 6), dpi=300)
plt.subplot(1, 2, 1)
plt.pcolor(input[1, 0])
plt.colorbar()

plt.subplot(1, 2, 2)
plt.pcolor(prediction[0, 0])
plt.colorbar()
```
